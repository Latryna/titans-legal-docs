```
import numpy as np
import networkx as nx
import time

# --- Helper Functions for Presentation ---
def print_header(title):
    """Prints a formatted header for each module's output."""
    print("\n" + "="*60)
    print(f"[[ {title} ]]")
    print("="*60)

def print_io(input_label, input_data, output_label, output_data):
    """Prints formatted input and output for a module."""
    print(f"  -> Input ({input_label}):\n     {input_data}")
    time.sleep(0.5) # Simulate processing time for dramatic effect
    print(f"  <- Output ({output_label}):\n     {output_data}")


# --- Module M1: Perception (CapsNet Placeholder) ---
class PerceptionModule:
    """
    Simulates M1. Takes raw sensory data and converts it into a structured vector.
    A real implementation would use a Capsule Network.
    """
    def process(self, raw_data: dict) -> np.ndarray:
        # In a real system, this would be a complex feature extraction process.
        # Here, we just combine values into a vector.
        vector = np.array([
            raw_data.get("feature1", 0.0),
            raw_data.get("feature2", 0.0),
            raw_data.get("urgency", 0.0)
        ])
        return vector

# --- Module M2: Memory (VAE Replay Placeholder) ---
class MemoryModule:
    """
    Simulates M2. Receives a vector and decides whether to store it.
    A real implementation would use a VAE for generative replay.
    """
    def __init__(self):
        self.short_term_memory = []

    def process(self, perception_vector: np.ndarray) -> str:
        # A simple "surprise" trigger based on vector magnitude.
        surprise_factor = np.linalg.norm(perception_vector)
        if surprise_factor > 1.0:
            self.short_term_memory.append(perception_vector)
            return f"High surprise ({surprise_factor:.2f}). Stored in STM."
        return f"Low surprise ({surprise_factor:.2f}). Ignored."

# --- Module M3: Abstraction (Transformer Placeholder) ---
class AbstractionModule:
    """
    Simulates M3. Takes a perception vector and maps it to a high-level semantic concept.
    A real implementation would use a Transformer Encoder to distill meaning.
    """
    def process(self, perception_vector: np.ndarray) -> str:
        # Rule-based mapping to a concept for demonstration.
        if perception_vector[2] > 0.8:
            return "CONCEPT_URGENT_TASK"
        if perception_vector[0] > perception_vector[1]:
            return "CONCEPT_ANALYTICAL_INPUT"
        return "CONCEPT_BACKGROUND_NOISE"

# --- Module M4: Reasoning (GAT Placeholder) ---
class ReasoningModule:
    """
    Simulates M4. Takes a concept and finds related concepts in a knowledge graph.
    A real implementation would use a Graph Attention Network (GAT).
    """
    def __init__(self):
        # Create a simple knowledge graph for the demo.
        self.knowledge_graph = nx.Graph()
        self.knowledge_graph.add_edges_from([
            ("CONCEPT_URGENT_TASK", "ACTION_ALLOCATE_RESOURCES"),
            ("CONCEPT_URGENT_TASK", "NOTIFY_SUPERVISOR"),
            ("CONCEPT_ANALYTICAL_INPUT", "ACTION_RUN_ANALYSIS"),
            ("ACTION_RUN_ANALYSIS", "SAVE_RESULTS")
        ])

    def process(self, concept: str) -> list:
        if self.knowledge_graph.has_node(concept):
            # Find all directly connected nodes (neighbors).
            related_concepts = list(self.knowledge_graph.neighbors(concept))
            return related_concepts
        return ["No related concepts found."]

# --- Module M5: Agency (Actor-Critic Placeholder) ---
class AgencyModule:
    """
    Simulates M5. Takes a list of potential actions and selects one.
    A real implementation uses a Bayesian Actor-Critic model with epistemic reward.
    """
    def process(self, potential_actions: list) -> str:
        # Simple decision logic: prioritize actions over notifications.
        for action in potential_actions:
            if "ACTION_" in action:
                return f"Decision: Execute '{action}'."
        if potential_actions:
            return f"Decision: Fallback to '{potential_actions[0]}'."
        return "Decision: No action required."


# --- Main Demonstration Function ---
def run_cognitive_trace(input_data: dict):
    """
    Runs the full simulation from M1 to M5 and prints the results.
    """
    print("="*60)
    print(">>> STARTING TITANS COGNITIVE TRACE SIMULATION <<<")
    print(f"Initial raw data: {input_data}")
    print("="*60)

    # Instantiate all modules
    m1_perception = PerceptionModule()
    m2_memory = MemoryModule()
    m3_abstraction = AbstractionModule()
    m4_reasoning = ReasoningModule()
    m5_agency = AgencyModule()

    # --- M1 ---
    print_header("M1: PERCEPTION")
    perception_vector = m1_perception.process(input_data)
    print_io("Raw Data", input_data, "Perception Vector", f"array({np.round(perception_vector, 2)})")

    # --- M2 ---
    print_header("M2: MEMORY")
    memory_status = m2_memory.process(perception_vector)
    print_io("Perception Vector", f"array({np.round(perception_vector, 2)})", "Memory Status", memory_status)

    # --- M3 ---
    print_header("M3: ABSTRACTION")
    abstract_concept = m3_abstraction.process(perception_vector)
    print_io("Perception Vector", f"array({np.round(perception_vector, 2)})", "Abstract Concept", f"'{abstract_concept}'")

    # --- M4 ---
    print_header("M4: REASONING")
    related_concepts = m4_reasoning.process(abstract_concept)
    print_io("Abstract Concept", f"'{abstract_concept}'", "Related Concepts/Actions", related_concepts)

    # --- M5 ---
    print_header("M5: AGENCY")
    final_decision = m5_agency.process(related_concepts)
    print_io("Potential Actions", related_concepts, "Final Decision", final_decision)

    print("\n" + "="*60)
    print(">>> SIMULATION COMPLETE <<<")
    print("="*60)


if __name__ == '__main__':
    # --- DEMO SCENARIO 1: An urgent task arrives ---
    print("\n\n--- SCENARIO 1: URGENT TASK ---")
    urgent_task_data = {"feature1": 0.5, "feature2": 0.3, "urgency": 0.9}
    run_cognitive_trace(urgent_task_data)

    # --- DEMO SCENARIO 2: A routine analytical input ---
    print("\n\n--- SCENARIO 2: ROUTINE ANALYSIS ---")
    analytical_data = {"feature1": 0.9, "feature2": 0.2, "urgency": 0.1}
    run_cognitive_trace(analytical_data)
```

```
# titans_step_logger.py
# Minimal cognitive-step logger (single-process, file-backed). Python 3.10+

import os, json, time, sqlite3, hashlib
from dataclasses import dataclass, asdict
from typing import Dict, Any, Optional
from pathlib import Path

DATA_DIR = Path("./data")
DATA_DIR.mkdir(exist_ok=True)
LOG_JSONL = DATA_DIR / "steps.jsonl"
DB_SQLITE = DATA_DIR / "index.sqlite"

# --- M1..M5 ultra-stubs (replace with real modules) ---
def m1_perception(x: str) -> Dict[str, Any]:
    # F: placeholder for real encoder/vision; here: toy features
    return {"tokens": x.split(), "len": len(x)}

def m3_abstraction(percept: Dict[str, Any]) -> Dict[str, Any]:
    # F: toy "concept": length bucket
    L = percept["len"]
    concept = "SHORT" if L < 64 else "LONG"
    return {"concept": concept}

def m4_reasoning(concept: Dict[str, Any]) -> Dict[str, Any]:
    # F: toy rule
    plan = "SUMMARIZE" if concept["concept"] == "LONG" else "VERBATIM"
    return {"plan": plan}

def m5_agency(plan: Dict[str, Any], x: str) -> Dict[str, Any]:
    # F: toy act
    if plan["plan"] == "SUMMARIZE":
        y = (x[:80] + "...") if len(x) > 80 else x
    else:
        y = x
    # F: toy uncertainty delta
    return {"action": plan["plan"], "output": y, "delta_uncertainty": 0.1 if plan["plan"]=="SUMMARIZE" else 0.02}

# --- Step schema ---
@dataclass
class Step:
    ts: float
    trace_id: str
    source: str
    event: str
    payload: Dict[str, Any]
    metrics: Dict[str, float]
    prev_hash: Optional[str] = None
    hash: Optional[str] = None

def step_hash(s: Step) -> str:
    b = json.dumps({
        "ts": s.ts, "trace_id": s.trace_id, "source": s.source, "event": s.event,
        "payload": s.payload, "metrics": s.metrics, "prev_hash": s.prev_hash
    }, sort_keys=True).encode("utf-8")
    return hashlib.sha256(b).hexdigest()

# --- Append-only log + index ---
def ensure_db():
    conn = sqlite3.connect(DB_SQLITE)
    c = conn.cursor()
    c.execute("""CREATE TABLE IF NOT EXISTS steps(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        ts REAL, trace_id TEXT, source TEXT, event TEXT,
        hash TEXT UNIQUE, prev_hash TEXT, offset INTEGER
    )""")
    c.execute("CREATE INDEX IF NOT EXISTS idx_trace ON steps(trace_id)")
    conn.commit(); conn.close()

def append_step(s: Step):
    # append JSONL
    line = json.dumps(asdict(s), ensure_ascii=False)
    with open(LOG_JSONL, "ab") as f:
        off = f.tell()
        f.write((line + "\n").encode("utf-8"))
    # index
    conn = sqlite3.connect(DB_SQLITE)
    c = conn.cursor()
    c.execute("INSERT INTO steps(ts, trace_id, source, event, hash, prev_hash, offset) VALUES(?,?,?,?,?,?,?)",
              (s.ts, s.trace_id, s.source, s.event, s.hash, s.prev_hash, off))
    conn.commit(); conn.close()

def latest_hash(trace_id: str) -> Optional[str]:
    conn = sqlite3.connect(DB_SQLITE); c = conn.cursor()
    c.execute("SELECT hash FROM steps WHERE trace_id=? ORDER BY id DESC LIMIT 1", (trace_id,))
    row = c.fetchone(); conn.close()
    return row[0] if row else None

# --- One cognitive step run ---
def run_step(trace_id: str, x: str):
    prev = latest_hash(trace_id)
    # M1
    p = m1_perception(x)
    s1 = Step(time.time(), trace_id, "M1", "PERCEPT", p, {"dt_ms": 0.0}, prev_hash=prev); s1.hash = step_hash(s1)
    append_step(s1)
    # M3
    a = m3_abstraction(p)
    s2 = Step(time.time(), trace_id, "M3", "ABSTRACT", a, {"dt_ms": 0.0}, prev_hash=s1.hash); s2.hash = step_hash(s2)
    append_step(s2)
    # M4
    r = m4_reasoning(a)
    s3 = Step(time.time(), trace_id, "M4", "REASON", r, {"dt_ms": 0.0}, prev_hash=s2.hash); s3.hash = step_hash(s3)
    append_step(s3)
    # M5
    act = m5_agency(r, x)
    s4 = Step(time.time(), trace_id, "M5", "ACT", act, {"dt_ms": 0.0, "delta_uncertainty": act["delta_uncertainty"]},
              prev_hash=s3.hash); s4.hash = step_hash(s4)
    append_step(s4)
    return act["output"]

if __name__ == "__main__":
    ensure_db()
    out = run_step(trace_id="demo-001", x="This is a very long input that should trigger the summarize plan in our toy pipeline.")
    print(out)
```

```
# -*- coding: utf-8 -*-
"""
TITANS Multi-Agent Orchestrator inspired by Semantic Kernel's Group Chat model.
"""
import textwrap
from typing import List, Dict, Tuple

# Assuming bootstrap_config is in the same package
# from .bootstrap_config import AGENT_ROLES, generate_system_prompt
# Import AGENT_ROLES and generate_system_prompt from the previous cell
from __main__ import AGENT_ROLES, generate_system_prompt

class TitansOrchestrator:
    """
    Acts as a "Group Chat Manager" to automate the TITANS multi-agent workflow.
    It maintains the project's state and routes tasks to the appropriate agent.
    """
    def __init__(self, agent_definitions: Dict):
        self.agents = list(agent_definitions.keys())
        self.chat_history: List[Tuple[str, str]] = []
        print("✅ TitansOrchestrator initialized. Agents onboarded:")
        for agent in self.agents:
            print(f"   - {agent}")
        print("-" * 40)

    def add_to_history(self, author: str, content: str):
        """Adds a message to the central chat history."""
        self.chat_history.append((author, content))

    def get_full_context(self) -> str:
        """Constructs the full, up-to-date context from history."""
        history_str = "\n".join([f"[{author}]: {content}" for author, content in self.chat_history])
        return f"""
        # --- CURRENT PROJECT HISTORY ---
        {history_str}
        # --- END OF HISTORY ---
        """

    def select_next_agent(self, task_description: str) -> str:
        """Selects the next agent based on task keywords."""
        task_lower = task_description.lower()
        # Prioritize Memory Management Agent if memory-related keywords are present
        if any(kw in task_lower for kw in ["memory", "storage", "recall", "episodic"]):
            return "New-Agent" # Return the key for the new agent in AGENT_ROLES
        elif any(kw in task_lower for kw in ["architect", "design", "synthesize", "vision", "decide"]):
            return "Gemini-Architect"
        elif any(kw in task_lower for kw in ["implement", "code", "secure", "test", "deploy", "fix"]):
            return "GPT-Agent"
        elif any(kw in task_lower for kw in ["document", "research", "write", "latex", "grant", "explain"]):
            return "Claude-Documenter"
        else:
            print("⚠️  Task is ambiguous. Defaulting to Architect.")
            return "Gemini-Architect"

    def execute_task(self, author: str, task: str):
        """Simulates a full orchestration cycle."""
        print(f"\n>>> New Task Received from [{author}]: '{task}'")
        self.add_to_history(author, task)
        next_agent = self.select_next_agent(task)
        print(f"🤖 Orchestrator routes task to: [{next_agent}]")

        system_prompt = generate_system_prompt(next_agent)
        full_context = self.get_full_context()

        final_prompt_for_agent = f"""
        {system_prompt}
        {full_context}

        ## 5. CURRENT TASK
        Based on all the context above, execute the following task: "{task}"
        """

        print(f"\n--- Injected Prompt for [{next_agent}] (simulation) ---")
        print(textwrap.shorten(final_prompt_for_agent, width=500, placeholder="..."))
        print("----------------------------------------------------")

        # In a real implementation, this would be an API call to the LLM.
        simulated_response = f"Acknowledged. Executing task: '{task}'. Deliverable will be a {AGENT_ROLES[next_agent]['deliverables'].split(',')[0]}."
        self.add_to_history(next_agent, simulated_response)
        print(f"✅ Task completed. History updated.")

if __name__ == "__main__":
    orchestrator = TitansOrchestrator(AGENT_ROLES)
    orchestrator.execute_task(
        author="Esu",
        task="Design the 'Homeostatic Core' module. Define its key variables and how it integrates with M5."
    )
    orchestrator.execute_task(
        author="Esu",
        task="Now, implement the 'Homeostatic Core' as a Python class. Include basic unit tests."
    )
    orchestrator.execute_task(
        author="Esu",
        task="Implement a short-term memory component for the Memory Management Agent."
    )
    orchestrator.execute_task(
        author="Esu",
        task="Design a long-term storage mechanism for the memory system."
    )
    print("\n\n" + "="*50)
    print("FINAL CONVERSATION HISTORY")
    print("="*50)
    print(orchestrator.get_full_context())
```

```
import json
import os

# Construct the full path to the steps.jsonl file
steps_file_path = "/content/data/steps.jsonl"

# Initialize an empty list to store the parsed JSON objects
cognitive_steps = []

# Open the steps.jsonl file for reading and parse each line as a JSON object
with open(steps_file_path, 'r') as f:
    for line in f:
        step = json.loads(line)
        cognitive_steps.append(step)

# At this point, the cognitive_steps list contains all parsed JSON objects
# You can inspect the list to verify the content
# print(cognitive_steps)
```

## Modify html to load data

### Subtask:
Add a script tag to the `titans_interactive_dashboard.html` file to fetch and load the JSON data.


**Reasoning**:
Add a script tag to the titans_interactive_dashboard.html file to fetch and load the JSON data.



```
import re

html_file_path = "titans_interactive_dashboard.html"

# Read the existing HTML file
with open(html_file_path, 'r') as f:
    html_content = f.read()

# Define the script to fetch the JSON data
fetch_script = """
    // Fetch and load cognitive trace data
    let cognitiveTraceData = null;
    fetch('/content/data/cognitive_data.json')
        .then(response => {
            if (!response.ok) {
                throw new Error(`HTTP error! status: ${response.status}`);
            }
            return response.json();
        })
        .then(data => {
            cognitiveTraceData = data;
            console.log('Cognitive trace data loaded successfully:', cognitiveTraceData);
            // You can now use cognitiveTraceData in other parts of your script
            // For example, call a function to update the dashboard with this data
            // updateDashboardWithTraceData(cognitiveTraceData);
        })
        .catch(error => {
            console.error('Error fetching cognitive trace data:', error);
        });
"""

# Locate the closing script tag
# We use a regex to find the last </script> tag before the closing </body>
# This assumes the main script is the last one before the body closes
script_end_tag_match = re.search(r'</script>\s*</body>', html_content, re.DOTALL)

if script_end_tag_match:
    # Insert the fetch script before the closing script tag
    insertion_point = script_end_tag_match.start()
    modified_html_content = html_content[:insertion_point] + fetch_script + "\n    " + html_content[insertion_point:]

    # Write the modified HTML content back to the file
    with open(html_file_path, 'w') as f:
        f.write(modified_html_content)

    print(f"Successfully added fetch script to {html_file_path}")
else:
    print("Could not find the main script tag to insert the fetch script.")
```
