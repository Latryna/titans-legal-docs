```
from dataclasses import dataclass
from typing import Dict, Any

@dataclass
class M7AffectiveEmpathyResonator:
    """Translates observed suffering into internal affective pain."""
    attenuation: float = 0.9

    def __post_init__(self):
        if not 0.0 <= self.attenuation <= 1.0:
            raise ValueError("Attenuation must be between 0 and 1.")

    def evaluate(self, observed_agent_state: Dict[str, Any]) -> Dict[str, float]:
        suffering = float(observed_agent_state.get("suffering_level", 0.0))
        affective_pain = max(0.0, min(1.0, suffering * self.attenuation))
        return {"affective_pain": affective_pain}
```
